# **Ecommerce Purchases**



**Dataset Link:** https://www.kaggle.com/datasets/utkarsharya/ecommerce-purchases

**Description:**
The Ecommerce Purchases dataset contains information about purchases made through an online ecommerce platform. Each row in the dataset represents a single purchase transaction, with various attributes recorded for analysis and tracking purposes. Below are the details of each column in the dataset:

1. **Address:** The address associated with the purchase delivery.
2. **Lot:** An identifier for the lot or batch of products associated with the purchase.
3. **AM or PM:** Indicates whether the purchase was made in the morning (AM) or the evening (PM).
4. **Browser Info:** Information about the web browser used by the customer during the purchase.
5. **Company:** The company from which the purchase was made, if applicable.
6. **Credit Card:** The credit card number used for the transaction (possibly masked for security).
7. **CC Exp Date:** The expiration date of the credit card used.
8. **CC Security Code:** The security code associated with the credit card used.
9. **CC Provider:** The provider or issuer of the credit card.
10. **Email:** The email address associated with the customer's account.
11. **Job:** The occupation or job title of the customer.
12. **IP Address:** The IP address of the device used for the purchase.
13. **Language:** The language preference of the customer.
14. **Purchase Price:** The price of the purchase transaction.

**Purpose:**
This dataset can be utilized for various analytical purposes, including but not limited to:

- Understanding customer demographics and behavior.
- Analyzing purchase patterns such as time of day preferences (AM/PM).
- Studying the distribution of purchases across different job roles or industries.
- Investigating the popularity of different web browsers or language preferences among customers.
- Analyzing the distribution of purchase prices and identifying trends in spending.
